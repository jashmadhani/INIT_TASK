```python
customers = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/customers.csv')
products = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/products.csv')
transactions = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/transactions.csv')
```

```python
print("\n Customers")
display(customers.head(3))

print("\n Products")
display(products.head(3))

print("\n Transactions ")
display(transactions.head(3))
```

```python
print("\n Missing Values ")
print("Customers:\n", customers.isnull().sum())
print("\nProducts:\n", products.isnull().sum())
print("\nTransactions:\n", transactions.isnull().sum())
```

```python
duplicate_tcount = transactions['transaction_id'].duplicated().sum()
print(f"\nDuplicate transaction_id entries: {duplicate_tcount}")
```

```python
transactions['timestamp'] = pd.to_datetime(transactions['timestamp'])
```

```python
earliest = transactions['timestamp'].min()
latest = transactions['timestamp'].max()

print(f"Date range of transactions: {earliest.date()} to {latest.date()}")
```

```python
transactions['hour_of_day'] = transactions['timestamp'].dt.hour
transactions['day_of_week'] = transactions['timestamp'].dt.dayofweek
transactions['month'] = transactions['timestamp'].dt.month
print(f"The Hour of the day, Day of week, month for the first 5 transactions are:")
display(transactions[['hour_of_day', 'day_of_week', 'month']].head())
```
